```python
from functools import cache
from typing import Tuple, Literal



table_row = Tuple[int, int, int, int, int, int]


@cache
def fib(n: int) -> int:
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)


def circ(tup: table_row) -> table_row:
    return tup[1:] + (tup[0] ^ (tup[1] * tup[2]),)


def special_func(n: int) -> int:
    if n == 1:
        return 1
    return fib(n + 2) - fib(n - 2)
```

```python
from itertools import product

res = 1


seen = set()

for initial_row in product((0,1), repeat=6):
    assert len(initial_row) == 6

    if initial_row in seen:
        continue

    print(initial_row)

    cur_row = circ(initial_row)
    circ_len = 1
    while cur_row != initial_row:

        print(f"\t{cur_row}")

        seen.add(cur_row)

        circ_len += 1
        cur_row = circ(cur_row)

    res *= special_func(circ_len)

print(res)
```
