```python
def sum_options(curr: list[int], seen: set[tuple], N: int) -> int:
    if len(curr) == 2 ** N:
        for i in range(1, N):
            if tuple(curr[-i:] + curr[:N - i]) in seen:
                return 0
            seen.add(tuple(curr[-i:] + curr[:N - i]))
        print("".join(map(str, curr)))
        return int("".join(map(str, curr)), 2)
    
    res = 0
    for next_bit in [0, 1]:
        if tuple(curr[-N + 1:] + [next_bit]) not in seen:
            next_list = curr.copy() + [next_bit]
            next_set = seen.copy() | {tuple(curr[-N + 1:] + [next_bit])}
            res += sum_options(next_list, next_set, N)
    return res 
```

```python
def S(N: int) -> int:
    return sum_options([0]* N, {tuple([0] * N)}, N)
```
